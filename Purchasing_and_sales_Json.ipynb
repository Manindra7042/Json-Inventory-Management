```python
import json
fd=open("record.json",'r')
r=fd.read()
fd.close()
record=json.loads(r)
record
sales={}
```

```python

for i in range(1,30):
  ui_prod=input("enter product id ")
  ui_qn=int(input("enter quantity"))
  print("********************************")
  print("Product:",record[ui_prod]["name"])
  print("Price:",record[ui_prod]["pr"])
  print("Billing amount",record[ui_prod]["pr"]*ui_qn)
  print("********************************")
  record[ui_prod]["qn"]=record[ui_prod]["qn"]-ui_qn
  sales[i]={"prod":ui_prod,"qn":ui_qn,"amount":record[ui_prod]["pr"]*ui_qn}
  sale=json.dumps(sales)
  fd=open('sales','w')
  fd.write(sale)
  fd.close()
```
